```python
import sqlalchemy 
import pymysql
import pandas as pd
df_1=pd.read_csv("C:\\Users\\VIGNESHWAR SANTHANAM\\Downloads\\OLA_DataSet.xlsx - July.csv")
connection = pymysql.connect(
    host="localhost",
    user="root",
    password="password",  # empty string
    database="zukun")
cursor = connection.cursor()

insert_query = """
        INSERT INTO Ride (Date,Time, Booking_ID, Booking_Status, Customer_ID, Vehicle_Type, 
     Pickup_Location, Drop_Location, V_TAT, C_TAT, Canceled_Rides_by_Customer, 
     Canceled_Rides_by_Driver, Incomplete_Rides, Incomplete_Rides_Reason, 
     Booking_Value, Payment_Method, Ride_Distance, Driver_Ratings, Customer_Rating, Vehicle_Images)
        VALUES (%s, %s, %s, %s, %s,%s,%s,%s,%s,%s,%s, %s, %s, %s, %s,%s,%s,%s,%s,%s)
        """
for i in range(df_1.shape[0]):
            row = df_1.iloc[i]
            cursor.execute(insert_query, (
        str(row["Date"])[:25],
        str(row["Time"])[:25],
        str(row["Booking_ID"])[:50],
        str(row["Booking_Status"])[:50],
        str(row["Customer_ID"])[:50],
        str(row["Vehicle_Type"])[:50],
        str(row["Pickup_Location"])[:100],
        str(row["Drop_Location"])[:100],
        None if pd.isna(row["V_TAT"]) else float(row["V_TAT"]),
        None if pd.isna(row["C_TAT"]) else float(row["C_TAT"]),
        None if pd.isna(row["Canceled_Rides_by_Customer"]) else str(row["Canceled_Rides_by_Customer"]),
        None if pd.isna(row["Canceled_Rides_by_Driver"]) else str(row["Canceled_Rides_by_Driver"]),
        None if pd.isna(row["Incomplete_Rides"]) else str(row["Incomplete_Rides"]),
        None if pd.isna(row["Incomplete_Rides_Reason"]) else str(row["Incomplete_Rides_Reason"]),
        int(row["Booking_Value"]),
        None if pd.isna(row["Payment_Method"]) else str(row["Payment_Method"]),
        int(row["Ride_Distance"]),
        None if pd.isna(row["Driver_Ratings"]) else float(row["Driver_Ratings"]),
        None if pd.isna(row["Customer_Rating"]) else float(row["Customer_Rating"]),
        str(row["Vehicle Images"])
    ))

# 6. Commit & close
connection.commit()
cursor.close()
connection.close()
```
